```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

1. Read an Image

```python
imgPath = 'images/messi.jpg'

imgColor = cv2.imread(imgPath, cv2.IMREAD_COLOR)
imgGray = cv2.imread(imgPath, cv2.IMREAD_GRAYSCALE)
imgUnchanged = cv2.imread(imgPath, cv2.IMREAD_UNCHANGED)
imgAnyColor = cv2.imread(imgPath, cv2.IMREAD_ANYCOLOR)
```

2. Display an image using OpenCV

```python
cv2.imshow('Image', imgColor)
cv2.imshow('Image', imgGray)
cv2.imshow('Image', imgUnchanged)
cv2.imshow('Image', imgAnyColor)

# Additional Functions
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
# Using Horizontal Stacking

cv2.namedWindow("main", cv2.WINDOW_NORMAL)
cv2.imshow('main', np.hstack((imgColor, imgColor, imgColor)))

cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
# Using Vertical Stacking

cv2.namedWindow("main", cv2.WINDOW_NORMAL)
cv2.imshow('main', np.vstack((imgColor, imgColor, imgColor)))

cv2.waitKey(5000)
cv2.destroyAllWindows()
```

3. Display an image using Matplotlib

```python
plt.imshow(imgColor)
```

```python
# Additional Functions
plt.figure(figsize=(4,4), num='test.img')   # width, height in inches

plt.imshow(imgColor)
plt.xticks([]), plt.yticks([])              # remove scaling
plt.show()
```

4. Display multiple images using Matplotlib

```python
plt.figure(figsize=(8,8), num='test.img')

img2 = cv2.imread(imgPath, 0)

# 01
plt.subplot(2,2,1)
plt.imshow(img2, cmap='Greys_r')
plt.xticks([]), plt.yticks([])
plt.title('Original')

# 02
plt.subplot(2,2,2)
plt.imshow(img2, cmap='Blues')
plt.xticks([]), plt.yticks([])
plt.title('Blue')

# 03
plt.subplot(2,2,3)
plt.imshow(img2, cmap='Greys')
plt.xticks([]), plt.yticks([])
plt.title('Gray Scale')

# 04
plt.subplot(2,2,4)
plt.imshow(img2, cmap='gist_heat')
plt.xticks([]), plt.yticks([])
plt.title('Gist Heat')
```

7. Changing Color space

```python
img = cv2.imread(imgPath, cv2.IMREAD_COLOR)

# BGR to Gray
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# BGR to HSV
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.figure(figsize=(10,10), num='test.img')

plt.subplot(1,2,1)
plt.imshow(img1)
plt.xticks([]), plt.yticks([])
plt.title('BGR to Gray')

plt.subplot(1,2,2)
plt.imshow(img2)
plt.xticks([]), plt.yticks([])
plt.title('BGR to HSV')
```
