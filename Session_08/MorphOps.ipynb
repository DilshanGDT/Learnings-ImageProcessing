A. Morphological Transformations

1. Create custom kernels for morphological operations

```python
# Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# kernel = cv2.getStructuringElement( cv2_shape, kernel_size )

# Rectangular Kernal
kernel1 = np.ones((5,5), np.uint8)

# Rectangular Kernal
kernel2 = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))

# Elliptical Kernal
kernel3 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))

# Cross-shaped Kernal
kernel4 = cv2.getStructuringElement(cv2.MORPH_CROSS, (5,5))
```

2. Erosion & Dilation

```python
img = cv2.imread(r'images\j.png', 0)
```

```python
# Erosion
erosion1 = cv2.erode(img, kernel1, iterations=1)
erosion2 = cv2.erode(img, kernel2, iterations=1)
erosion3 = cv2.erode(img, kernel3, iterations=1)
erosion4 = cv2.erode(img, kernel4, iterations=1)
```

```python
titles = ['Original Image', 'Erosion 1', 'Erosion 2', 'Erosion 3', 'Erosion 4',]
images = [img, erosion1, erosion2, erosion3, erosion4]

plt.figure(figsize=(13,13), num='test.img')

for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.title(titles[i], fontsize = 8)
    plt.xticks([]), plt.yticks([])
    plt.imshow(images[i], 'gray')
```

```python
# Dilation
dilation1 = cv2.dilate(img, kernel1, iterations=1)
dilation2 = cv2.dilate(img, kernel2, iterations=1)
dilation3 = cv2.dilate(img, kernel3, iterations=1)
dilation4 = cv2.dilate(img, kernel4, iterations=1)
```

```python
titles = ['Original Image', 'Diation 1', 'Diation 2', 'Diation 3', 'Diation 4',]
images = [img, dilation1, dilation2, dilation3, dilation4]

plt.figure(figsize=(13,13), num='test.img')

for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.title(titles[i], fontsize = 8)
    plt.xticks([]), plt.yticks([])
    plt.imshow(images[i], 'gray')
```

3. Opening & Closing

```python
# Opening
opening1 = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel1)
opening2 = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel2)
opening3 = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel3)
opening4 = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel4)
```

```python
titles = ['Original Image', 'Opening 1', 'Opening  2', 'Opening  3', 'Opening  4',]
images = [img, opening1, opening2, opening3, opening4]

plt.figure(figsize=(13,13), num='test.img')

for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.title(titles[i], fontsize = 8)
    plt.xticks([]), plt.yticks([])
    plt.imshow(images[i], 'gray')
```

```python
# Closing
closing1 = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel1)
closing2 = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel2)
closing3 = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel3)
closing4 = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel4)
```

```python
titles = ['Original Image', 'Closing 1', 'Closing  2', 'Closing  3', 'Closing  4',]
images = [img, closing1, closing2, closing3, closing4]

plt.figure(figsize=(13,13), num='test.img')

for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.title(titles[i], fontsize = 8)
    plt.xticks([]), plt.yticks([])
    plt.imshow(images[i], 'gray')
```

B. Contours

###### Contours can be explained as a curve joining all the continuous points (along the boundary), having same color or intensity. 
###### The contours are a useful tool for shape analysis and object detection and recognition.

Drawing Contours

```python
img = cv2.imread(r'images\plus.jpg', 1)

imgCopy1 = img.copy()
imgCopy2 = img.copy()
imgCopy3 = img.copy()
```

```python
img_g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(img_g, 50, 255, 0)

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# resultImg = cv2.drawContours(srcImg, contours, index, borderColor, lineWidth)

img1 = cv2.drawContours(imgCopy3, contours, -1, (0,255,0), 2)
img2 = cv2.drawContours(imgCopy1, contours,  3, (255,0,0), 2)

cnt1 = contours[5]
cnt2 = contours[6]

# resultImg = cv2.drawContours(imgCopy, [cnt], 0, (0,0,255), 2)

img3 = cv2.drawContours(imgCopy2, [cnt1, cnt2], -1, (0,0,255), 2)
```

```python
titles = ['Original Image', 'All Contours', '4th Contour', '6th & 7th Contour']
images = [img, img1, img2, img3]

plt.figure(figsize=(13,13), num='test.img')

for i in range(4):
    plt.subplot(1, 5, i+1)
    plt.title(titles[i], fontsize = 8)
    plt.xticks([]), plt.yticks([])
    plt.imshow(images[i], 'gray')
```
