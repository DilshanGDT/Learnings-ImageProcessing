2. Image Resizing

```python
img = cv2.imread('images/flower.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# img.shape --> rows, cols --- height, width

# keep aspect-ratio
# Image resizing percentage
scalePercent = 25               # 200

# calculate the 25% of the original dimensions
# width - int(img, shape[1]/4) or 
width = int(img.shape[1] * scalePercent/100)
height = int(img.shape[0] * scalePercent/100)

resized = cv2.resize(img, dsize=(width, height))

print('image shape : ', img.shape)
print('resized image shape : ', resized.shape)
```

```python
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original')

plt.subplot(1,2,2)
plt.imshow(resized)
plt.title('Resized Image')
```

3. Splitting & merging

```python
img = cv2.imread('images/flower.jpg', cv2.IMREAD_COLOR)
img_ = img.copy()

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

Splitting

```python
# splitting image into 3 seperate B G R layers
b, g, r = cv2.split(img)
```

```python
plt.figure(figsize=(10,10), num='test.img')

# Single channel consider as greyscale images
plt.subplot(1,3,1)
plt.imshow(b, cmap='grey')
plt.xticks([]), plt.yticks([])
plt.title('Blue Channel')

plt.subplot(1,3,2)
plt.imshow(g, cmap='grey')
plt.xticks([]), plt.yticks([])
plt.title('Green Channel')

plt.subplot(1,3,3)
plt.imshow(r, cmap='grey')
plt.xticks([]), plt.yticks([])
plt.title('Red Channel')
```

Merging

```python
# remove blue layer - masking all values to become 0
b_ = b.copy()
b_[ :, :] = 0 # or r[ : ]

# remove red layer - masking all values to become 0
r_ = r.copy()
r_[ :, :] = 0 # or r[ : ]

# remove green layer - masking all values to become 0
g_ = g.copy()
g_[ :, :] = 0 # or r[ : ]
```

```python
# merge layers
mImgBlueRemoved = cv2.merge((b_, g, r))
mImgRedRemoved = cv2.merge((b, g, r_))
mImgGreenRemoved = cv2.merge((b, g_, r))

img_b = cv2.cvtColor(mImgBlueRemoved, cv2.COLOR_BGR2RGB)
img_r = cv2.cvtColor(mImgRedRemoved, cv2.COLOR_BGR2RGB)
img_g = cv2.cvtColor(mImgGreenRemoved, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(12,12), num='test.img')

plt.subplot(1, 4, 1)
plt.xticks([]), plt.yticks([])
plt.title('Original', fontsize= 8)
plt.imshow(img)

plt.subplot(1, 4, 2)
plt.xticks([]), plt.yticks([])
plt.title('Red Removed', fontsize= 8)
plt.imshow(img_r)

plt.subplot(1, 4, 3)
plt.xticks([]), plt.yticks([])
plt.title('green Removed', fontsize= 8)
plt.imshow(img_g)

plt.subplot(1, 4, 4)
plt.xticks([]), plt.yticks([])
plt.title('blue Removed', fontsize= 8)
plt.imshow(img_b)
```

4. Arithmatic operations

```python
img4 = cv2.imread(r'images/scr3.jpg', cv2.IMREAD_GRAYSCALE)
img5 = cv2.imread(r'images/scr4.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Addition
add = cv2.add(img4, img5)

print('img_4 pixel 1 : ', img4[1,1])
print('img_5 pixel 1 : ', img5[1,1])
print('add pixel 1 : ', add[1,1])

plt.figure(figsize=(12,12), num='test.img')

plt.subplot(1, 3, 1)
plt.xticks([]), plt.yticks([])
plt.title('Image 4', fontsize= 8)
plt.imshow(img4)

plt.subplot(1, 3, 2)
plt.xticks([]), plt.yticks([])
plt.title('Image 5', fontsize= 8)
plt.imshow(img5)

plt.subplot(1, 3, 3)
plt.xticks([]), plt.yticks([])
plt.title('Image 4 + image 5', fontsize= 8)
plt.imshow(add)
```

```python
# Weighted Addition

add_w = cv2.addWeighted(img4, 0.3, img5, 0.5, 0)

print('img_1 pixel 1 : ', img4[1,1])
print('img_2 pixel 1 : ', img5[1,1])
print('add pixel 1 : ', add_w[1,1])

plt.figure(figsize=(12,12), num='test.img')

plt.subplot(1, 3, 1)
plt.xticks([]), plt.yticks([])
plt.title('Image 4', fontsize= 8)
plt.imshow(img4)

plt.subplot(1, 3, 2)
plt.xticks([]), plt.yticks([])
plt.title('Image 5', fontsize= 8)
plt.imshow(img5)

plt.subplot(1, 3, 3)
plt.xticks([]), plt.yticks([])
plt.title('Image 4 + image 5 with weight', fontsize= 8)
plt.imshow(add_w)
```

```python
# Substraction
substract = cv2.subtract(img4, img5)

# Multiplication
multiply = cv2.multiply(img4, img5)

# Division
division = cv2.divide(img4, img5)
```

```python
plt.figure(figsize=(12,12), num='test.img')

plt.subplot(1, 5, 1)
plt.xticks([]), plt.yticks([])
plt.title('Image 4', fontsize= 8)
plt.imshow(img4)

plt.subplot(1, 5, 2)
plt.xticks([]), plt.yticks([])
plt.title('Image 5', fontsize= 8)
plt.imshow(img5)

plt.subplot(1, 5, 3)
plt.xticks([]), plt.yticks([])
plt.title('Image 4 - image 5', fontsize= 8)
plt.imshow(substract)

plt.subplot(1, 5, 4)
plt.xticks([]), plt.yticks([])
plt.title('Image 4 x image 5', fontsize= 8)
plt.imshow(multiply)

plt.subplot(1, 5, 5)
plt.xticks([]), plt.yticks([])
plt.title('Image 4 / image 5', fontsize= 8)
plt.imshow(division)
```

5. Logical Operations

```python
img6 = cv2.imread(r'images/flower.jpg', cv2.IMREAD_GRAYSCALE)
img7 = cv2.imread(r'images/mask_single.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# AND
and_ = cv2.bitwise_and(img6, img7)

# OR
or_ = cv2.bitwise_or(img6, img7)

# NOT
not_ = cv2.bitwise_not(img6, img7)

# XOR
xor_ = cv2.bitwise_xor(img6, img7)
```

```python
plt.figure(figsize=(10,6), num='test.img')

plt.subplot(2, 3, 1)
plt.xticks([]), plt.yticks([])
plt.title('Image 6', fontsize= 8)
plt.imshow(img6)

plt.subplot(2, 3, 2)
plt.xticks([]), plt.yticks([])
plt.title('Image 7', fontsize= 8)
plt.imshow(img7)

plt.subplot(2, 3, 3)
plt.xticks([]), plt.yticks([])
plt.title('Image 6 AND image 7', fontsize= 8)
plt.imshow(and_)

plt.subplot(2, 3, 4)
plt.xticks([]), plt.yticks([])
plt.title('Image 6 OR image 7', fontsize= 8)
plt.imshow(or_)

plt.subplot(2, 3, 5)
plt.xticks([]), plt.yticks([])
plt.title('Image 6 NOT image 7', fontsize= 8)
plt.imshow(not_)

plt.subplot(2, 3, 6)
plt.xticks([]), plt.yticks([])
plt.title('Image 6 XOR image 7', fontsize= 8)
plt.imshow(xor_)
```

6. Masking

```python
# A mask image is simply an image where some of the pixel intensity values are zero, and others are non-zero.
# The key point of masks is that they allow us to focus our computation only on regions of the image that interest us.
```

```python
# creating a black mask (single channel)
```

```python
img6 = cv2.imread(r'images/flower.jpg', cv2.IMREAD_COLOR)
img7 = cv2.imread(r'images/mask_single.jpg', cv2.IMREAD_COLOR)

and_ = cv2.bitwise_and(img6, img7)
```

```python
plt.figure(figsize=(10,10), num='test.img')

plt.subplot(1, 3, 1)
plt.xticks([]), plt.yticks([])
plt.title('Image 6', fontsize= 8)
plt.imshow(img6)

plt.subplot(1, 3, 2)
plt.xticks([]), plt.yticks([])
plt.title('Image 7', fontsize= 8)
plt.imshow(img7)

plt.subplot(1, 3, 3)
plt.xticks([]), plt.yticks([])
plt.title('Image 6 AND image 7 = Apply Mask', fontsize= 8)
plt.imshow(and_)
```
